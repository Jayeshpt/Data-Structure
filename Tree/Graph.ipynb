```python
class Graph:
    def __init__(self, directed=False):
        self.graph = {}
        self.directed = directed

    def add_node(self, data):
        if data in self.graph:
            print(data, "is already present in the graph")
        else:
            self.graph[data] = []

    def add_edge(self, v1, v2):
        if v1 not in self.graph or v2 not in self.graph:
            print("One or both of the nodes do not exist in the graph")
        else:
            self.graph[v1].append(v2)
            if not self.directed:
                self.graph[v2].append(v1)

    def delete_node(self, data):
        if data not in self.graph:
            print(data, "is not present in the graph")
            return
        del self.graph[data]
        for node in self.graph:
            self.graph[node] = [i for i in self.graph[node] if i != data]

    def delete_edge(self, v1, v2):
        if v1 not in self.graph or v2 not in self.graph:
            print("One or both of the nodes do not exist in the graph")
            return
        self.graph[v1] = [i for i in self.graph[v1] if i != v2]
        if not self.directed:
            self.graph[v2] = [i for i in self.graph[v2] if i != v1]

    def dfs(self, start_node):
        visited = set()
        stack = [start_node]
        while stack:
            node = stack.pop()
            if node not in visited:
                visited.add(node)
                print(node, end=" ")
                for neighbor in self.graph[node]:
                    if neighbor not in visited:
                        stack.append(neighbor)

    def bfs(self, start_node):
        visited = set()
        queue = [start_node]
        visited.add(start_node)
        while queue:
            node = queue.pop(0)
            print(node, end=" ")
            for neighbor in self.graph[node]:
                if neighbor not in visited:
                    visited.add(neighbor)
                    queue.append(neighbor)

    def search(self, data):
        return data in self.graph

# Usage example
gp = Graph()
gp.add_node("A")
gp.add_node("B")
gp.add_node("C")
gp.add_node("D")
gp.add_node("E")
gp.add_node("F")
gp.add_node("G")

gp.add_edge("A", "B")
gp.add_edge("A", "C")
gp.add_edge("B", "D")
gp.add_edge("C", "B")
gp.add_edge("A", "E")
gp.add_edge("F", "G")
gp.add_edge("C", "F")

print(gp.search("A"))

# gp.dfs("A")
# print()
gp.bfs("A")
print()

print(gp.search("G"))
```
