```python
class Bst:
    def __init__(self, key):
        self.key = key
        self.lchild = None
        self.rchild = None

    def insert(self, data):
        if self.key is None:
            self.key = data
        if self.key == data:
            return
        if self.key > data:
            if self.lchild:
                self.lchild.insert(data)
            else:
                self.lchild = Bst(data)
        else:
            if self.rchild:
                self.rchild.insert(data)
            else:
                self.rchild = Bst(data)

    def search(self, data):
        if self.key == data:
            print("Element found in the tree.")
            return
        if self.key > data:
            if self.lchild:
                self.lchild.search(data)
            else:
                print("Element is not found.")
        else:
            if self.rchild:
                self.rchild.search(data)
            else:
                print("Element is not found.")

    def pre_order(self):
        print(self.key, end=" ")
        if self.lchild:
            self.lchild.pre_order()
        if self.rchild:
            self.rchild.pre_order()

    def find_min(self):
        if not self.lchild:
            return self.key
        return self.lchild.find_min()

    def delete(self, data):
        if self.key is None:
            print("Tree is empty.")
        if data < self.key and self.lchild:
            self.lchild = self.lchild.delete(data)
        elif data > self.key and self.rchild:
            self.rchild = self.rchild.delete(data)
        else:
            if not self.lchild:
                return self.rchild
            elif not self.rchild:
                return self.lchild
            else:
                min_val = self.rchild.find_min()
                self.key = min_val
                self.rchild = self.rchild.delete(min_val)
        return self


bstt = Bst(10)
list1 = [1,2]

for i in list1:
    bstt.insert(i)
bstt.pre_order()
bstt.delete(10)
print()
bstt.pre_order()
```

```python
class heap:
    def __init__(self):
        self.heap = []
        self.size = 0
    def parent(self,i):
        return (i-1)//2
    def left_child(self,i):
        return 2 * i +1
    def right_child(self,i):
        return 2 * i + 2
    def swap (self,i,j):
        self.heap[i],self.heap[j] = self.heap[j],self.heap[i]
    def insert(self,data):
        self.heap.append(data)
        self.size +=1
        self.heapify_up(self.size-1)
    def heapify_up(self,i):
        while i>0 and self.heap[i]<self.heap[self.parent(i)]:
            self.swap(i,self.parent(i))
            i = self.parent(i)
            
    def Remove_min(self):
        min_value = self.heap[0]
        self.heap[0] = self.heap[self.size-1]
        self.heap.pop()
        self.size -=1
        self.heapify_down(0,self.size)
        return min_value
    def heapify_down(self,i,n):
        min_index = i
        right = right_child(i)
        left = left_child(i)
        
        while left < n and self.heap[left] < self.heap[min_index]:
            min_index = left
        while right < n and self.heap[right] < self.heap[min_index]:
            min_index = right
        if i != min_index:
            self.swap(i,min_index)
            self.heapify_down(min_index,n)
    def heap_sort(self):
        sorted_arr = []
        while self.size > 0:
            sorted_arr.append(self.Remove_min())
        return sorted_arr
    
    def delete(self,value):
        index = self.heap.index(value)
        self.heap[index],self.heap[-1]=self.heap[-1],self.heap[index]
        self.heap.pop()
        self.heapify_up(index)
        
        
            
            
```
