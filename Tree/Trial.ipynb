```python
class TreeNode:
    def __init__(self,data):
        self.data=data
        self.children = []
        self.parent=None
    
    def add_child(self,child):
        child.parent=self
        self.children.append(child)
        
    def get_level(self):
        level=0
        p=self.parent
        while p:
            level+=1
            p=p.parent
        return level
    
    def print_Tree(self):
        spaces = " " * self.get_level() * 2
        prefix = spaces + "|__" if self.parent else ""
        print(prefix + self.data)
        
        if self.children:
            for child in self.children:
                child.print_Tree()
                
def construct_tree():
    root = TreeNode("Electronics")
    
    laptop = TreeNode("Laptops")
    laptop.add_child(TreeNode("Mac"))
    laptop.add_child(TreeNode("Realme"))
    laptop.add_child(TreeNode("Asus"))
    laptop.add_child(TreeNode("MSI"))
    
    
    mobile = TreeNode("Mobiles")
    mobile.add_child(TreeNode("Iphone"))
    mobile.add_child(TreeNode("samsung"))
    mobile.add_child(TreeNode("Redmi"))
    
    root.add_child(laptop)
    root.add_child(mobile)
    
    root.print_Tree()
    
    
if __name__ == "__main__":
    construct_tree()
    
```
