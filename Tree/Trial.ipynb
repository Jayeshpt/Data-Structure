```python
class TreeNode:
    def __init__(self,data):
        self.data = data
        self.children = []
        self.parent = None
        
    def add_child(self,child):
        child.parent= self
        self.children.append(child)
        
    def get_level(self):
        level =0
        p = self.parent
        while p:
            level +=1
            p = p.parent
        return level
    def print_tree(self):
        spaces = " " * self.get_level() * 3
        prefix = spaces + "|__" if self.parent else ""
        print (prefix + self.data)
        if self.children:
            for child in self.children:
                child.print_tree()
                
                
def construct_Tree():
    root = TreeNode("Electronic")
    
    mobile = TreeNode("Mobiles")
    mobile.add_child(TreeNode("samsung"))
    mobile.add_child(TreeNode("realme"))
    mobile.add_child(TreeNode("apple"))
    
    
    earphone = TreeNode("Earphone")
    earphone.add_child(TreeNode("Noise"))
    earphone.add_child(TreeNode("Oneplus"))
    earphone.add_child(TreeNode("Earpod"))
    
    
    root.add_child(mobile)
    root.add_child(earphone)
    
    root.print_tree()
    
    
    
if __name__ =="__main__":
    construct_Tree()
```

## Binary search Tree


```python
class Binary_search_Tree:
    def __init__(self,key):
        self.key = key
        self.lchild = None
        self.rchild = None
        
    def insertion(self,data):
        if self.key is None:
            self.key = data
            return
        if self.key == data:
            return
        if self.key > data:
            if self.lchild:
                self.lchild.insertion(data)
            else:
                self.lchild = Binary_search_Tree(data)
        else:
            if self.rchild:
                self.rchild.insertion(data)
            else:
                self.rchild = Binary_search_Tree(data)
                
    def search(self,data):
        if self.key == data:
            print("element is found")
            return
        if self.key > data:
            if self.lchild:
                self.lchild.search(data)
                
            else:
                print('element is not found')
        else :
            if self.rchild:
                self.rchild.search(data)
            else:
                print("element is not found")
                
                
                
root = Binary_search_Tree(10)
datas = [1,2,3,4,5,6]

for i in datas:
    root.insertion(i)
    
root.search(6)
root.search(7)
```
